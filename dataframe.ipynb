## What is Dataframe

- 多組共享 Index 的 Series 組合為一個具備列索引（row index）與欄標籤（column label）的資料集
- 典型Table like的資料結構，操作相對直覺方便


```python
import numpy as np
import pandas as pd
```

```python
# make a self-defined dataframe

student_score = [100, 90, 80, 70]
student_name = ['A', 'B', 'C', 'D']
```

```python
df = pd.DataFrame()
df["student_name"] = student_name
df["student_score"] = student_score

print(df)
print(df.index)
print(df.columns)
print(df.student_score)
```
