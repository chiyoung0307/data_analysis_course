```python
import pandas as pd

# read data from webpage, this will load entire html form, return list of dataframe
df_list = pd.read_html('https://pandas.pydata.org/docs/user_guide/io.html')
```

```python
print(type(df_list))

print(df_list)
```

```python
df = df_list[0]
df
```

- 上面表格其實就是幾個pandas提供的df io
- 常用的通常是read_csv, read_json, read_html, read_excel
- 有時有輸出成其它格式需求，就使用對應的Writer, 常用有： to_csv, to_excel, to_json

```python
# export dataframe to a csv file

df.to_csv('df_io.csv')
# check your files from current working directory
```

```python
# if you don't need index, add parameter index=False

df.to_csv('df_io.csv', index=False)
```

```python
# read the file you just exported
pd.read_csv('df_io.csv')
```
