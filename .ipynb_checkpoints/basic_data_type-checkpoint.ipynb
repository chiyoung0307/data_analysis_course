### Python的資料型態
- 以下列出常見的，其實不只
- 又安裝第三方套件後，可能有更多不同種類資料型態，例如安裝pandas後：
    - Series
    - Data Frame
 


```python
# type()檢查變數的資料形態
x = 5
print(type(x))
```

```python
x = "Hello World"
print(type(x))
```

```python
x = 20.5
print(type(x))
```

```python
x = ["apple", "banana", "cherry"]
print(type(x))
```

```python
x = ("apple", "banana", "cherry")
print(type(x))
```

```python
x = {"name" : "John", "age" : 36}
print(type(x))
```
